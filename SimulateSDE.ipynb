```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from SimulateSDE import SDE

%load_ext autoreload

%autoreload 2
```

```python
T_0 = 0
T = 1
p = 0.1
N = 100
delta_t = 0.01
nb_simu = 20000
```

```python
n = np.random.binomial(N, p)
t = [0] + np.random.uniform(0, 1, n)
t = np.around(t, decimals=2)
```

```python
params_BS = {'mu' : 2, "sigma":0.3}
BS = SDE("BS", 1, params_BS)
BS_simu = BS.BlackScholes(nb_simu=nb_simu)
BS_simu = BS.plot_df(X=BS_simu, t=t, nb_simu=nb_simu)
BS_simu = BS.save_csv(X=BS_simu, t=list(t))
```

```python
params_OU = {"k" : 2, "m":4, "sigma":0.3}
X_OU = SDE("OU", 1, params_OU)
X_OU_simu = X_OU.plot_df(t=t, nb_simu=nb_simu)
X_OU_simu = X_OU.save_csv(X=X_OU_simu, t=t)
```

```python
params_Heston = {"mu" : 2, "sigma" : 0.3, 'v0' :4, 'k':2, 'm':4}
X_Heston = SDE("Heston", 1, params_Heston)
X_Heston_simu = X_Heston.plot_df(t=t, nb_simu=nb_simu)
X_Heston_simu = X_Heston.save_csv(X=X_Heston_simu, t=t)
```
